```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/train.csv")
```

```python
data["price_range"].unique()
```

```python
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 15))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for i, col in enumerate(data.columns):
    row_index = i // 5
    col_index = i % 5
    data[col].plot(kind='box', ax=axes[row_index, col_index])
    axes[row_index, col_index].set_title(col)

plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(20,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Data')
plt.show()
```

```python
median_value = data['fc'].median()
data['fc'] = np.where((data['fc'] > 13), median_value, data['fc'])
```

```python
data['fc'].plot(kind='box', figsize=(10, 5))
plt.title('Box Plot of Data[fc]')
plt.show()
```

```python
median_value = data['px_height'].median()
data['px_height'] = np.where((data['px_height'] > 1900), median_value, data['px_height'])
```

```python

data['px_height'].plot(kind='box', figsize=(10, 5))
plt.title('Box Plot of Data[px_height]')
plt.show()
```

```python
# prompt: decision tree moe

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop('price_range', axis=1), data['price_range'], test_size=0.2, random_state=42)

# Train the decision tree model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python

from sklearn import svm

clf_svm = svm.SVC(kernel='linear', random_state=42)

clf_svm.fit(X_train, y_train)

y_pred_svm = clf_svm.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred_svm))
```

```python


from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(n_estimators=11, random_state=42)

clf_rf.fit(X_train, y_train)

y_pred_rf = clf_rf.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred_rf))
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# prompt: knn model

from sklearn.neighbors import KNeighborsClassifier

# Create a KNN model with k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model to the training data
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred_knn = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_knn)
print('Accuracy:', accuracy)
```
